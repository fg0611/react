```python
# Función para convertir a días
def convertir_a_dias(texto):
    if not isinstance(texto, str):
        return 0
    
    texto = texto.lower()  # Convertir a minúsculas para manejar mayúsculas y minúsculas
    
    # Extraer el número usando una expresión regular
    numero = int(re.search(r'\d+', texto).group())
    
    # Determinar la unidad de tiempo
    if "día" in texto or "dia" in texto or "day" in texto:
        return numero
    elif "semana" in texto or "semanas" in texto or "week" in texto:
        return numero * 7
    elif "mes" in texto or "month" in texto:
        return numero * 30
    else:
        return 0  # Si no se reconoce el formato

# Aplicar la función a la columna "published"
# df["dias"] = df["published"].apply(convertir_a_dias)

# print(df)
```

```python
def fix_url(url):
    # Verificar si la URL contiene 'linkedin.com/jobs/view/'
    urlMatch = "linkedin.com/jobs/view/"
    # Expresión regular para extraer el ID
    match = re.search(r'/view/.*?-(\d+)(\?|$)', url)
    # Verificar si se encontró un match
    if match:
        id = match.group(1)  # Extraer el ID
        return f'{urlMatch}{id}'
```

```python
df["url"] = df["url"].apply(fix_url)
```

```python
import unicodedata
from scripts.csv_handler import csv_to_df
# from scripts.helpers import clean_text


def clean_text(text):
    """Removes problematic characters and replaces them with standard ones."""
    if isinstance(text, str):
        # Normalize Unicode characters (e.g., ’ → ')
        text = unicodedata.normalize("NFKD", text)

        # Use explicit Unicode escape sequences to replace problematic characters
        text = re.sub(r"[\u2018\u2019\u00B4\u0060]", "'", text)  # Replaces all special apostrophes (‘ ’ ´ `) with standard '
        text = re.sub(r"[\u201C\u201D\u201E]", '"', text)  # Replaces all special double quotes (“ ” „) with standard "
        text = re.sub(r"[\u2013\u2014\u2212]", "-", text)  # Replaces different types of dashes (– — −) with a standard hyphen (-)

        # Remove any non-printable characters
        text = "".join(char for char in text if char.isprintable())

        # Replace multiple spaces with a single space
        text = re.sub(r"\s+", " ", text).strip()

    return text

filename = 'jobs-webdeveloper2.csv'
df = csv_to_df(filename)
# df = df.apply(clean_text)
# Apply the function to all string values in the DataFrame
df["description"] = df["description"].astype(str).apply(clean_text)
df.to_csv(filename, index=False, encoding="utf-8")
```
